```python
def levantarStream(df, orden_meses, comuna):
    # Crear el gráfico de tipo stream con Plotly Express
    fig = px.area(df, x='mes', y='tasa', color='delito',
                  title=f'Streamgraph de Tasas de Delitos en {comuna}',
                  labels={'tasa': 'Tasa de Delito cada/100.000 hab.', 'delito': 'Tipo de Delito'},
                  category_orders={"mes": orden_meses})
    
    
    # Agregar líneas verticales y una etiqueta para Verano
    fig.add_shape(dict(type='line', x0='Enero', x1='Enero', y0=0, y1=1, xref='x', yref='paper',
                       line=dict(color='red', width=2, dash='dash')))
    fig.add_shape(dict(type='line', x0='Marzo', x1='Marzo', y0=0, y1=1, xref='x', yref='paper',
                       line=dict(color='red', width=2, dash='dash')))
    fig.add_annotation(text='Verano', x='Febrero', y=-2.5, showarrow=False, font=dict(color='red'))
    
    
    #Linea y etiqueta otoño
    fig.add_shape(dict(type='line', x0='Junio', x1='Junio', y0=0, y1=1, xref='x', yref='paper',
                       line=dict(color='green', width=2, dash='dash')))
    fig.add_annotation(text='Otoño', x=3.5, y=-2.5, showarrow=False, font=dict(color='green'))
    
    #Linea y etiqueta Invierno
    fig.add_shape(dict(type='line', x0='Septiembre', x1='Septiembre', y0=0, y1=1, xref='x', yref='paper',
                       line=dict(color='blue', width=2, dash='dash')))
    fig.add_annotation(text='Invierno', x=6.5, y=-2.5, showarrow=False, font=dict(color='blue'))
    
    #Linea y etiqueta Primavera
    fig.add_shape(dict(type='line', x0='Diciembre', x1='Diciembre', y0=0, y1=1, xref='x', yref='paper',
                       line=dict(color='#FFCC0F', width=2, dash='dash')))
    fig.add_annotation(text='Primavera', x=9.5, y=-2.5, showarrow=False, font=dict(color='#FFCC0F'))
    
    
    # Mostrar el gráfico
    fig.show()
```

```python
colores = {
    "Robo con Violencia o Intimidación": ["#FF0000", "#0000FF"],
    "Robo Lugar Habitado": ["#FFFF00", "#008000"],
    "Hurtos": ["#AC6F00", "#800080"]
}
```

```python
def deployComparacion(delito, df1, df2, maximo):
    #print(maximos)
    
    # Filtrar por delito especifico
    df_del1 = df1[df1["delito"]== delito]
    df_del2 = df2[df2["delito"]== delito]

    nombre1 = df_del1["nombre"].unique()
    nombre2 = df_del2["nombre"].unique()

    meses = df_del1["mes"].unique()
    
    # Extraer las tres primeras letras de cada mes
    meses_abreviados = [mes[:3] for mes in meses]

    # Ajustar las posiciones de las barras
    ancho_barra = 0.35

    pos_del1 = np.arange(len(meses))
    pos_del2 = pos_del1 + ancho_barra
    
    # Configurar el gráfico de barras agrupadas por mes
    plt.bar(pos_del1, df_del1["tasa"], width=ancho_barra, label=f"{nombre1[0]}", color='gray', edgecolor='black')
    plt.bar(pos_del2, df_del2["tasa"], width=ancho_barra, label=f"{nombre2[0]}", alpha=0.7, color='lightgray', edgecolor='black')

    # Añadir etiquetas y título
    plt.xlabel("Mes")
    plt.ylabel("Tasa de Delito")
    plt.title(f"Comparación Mensual de la Tasa de {delito} entre {nombre1[0]} y {nombre2[0]}")
    plt.xticks(pos_del1 + ancho_barra / 2, meses_abreviados)  # Etiquetas en el centro de las agrupaciones

    # Añadir líneas verticales para identificar las estaciones con etiquetas
    plt.axvline(x=2.7, color='black', linestyle='--', alpha=0.9)  # Entre Febrero y Marzo
    plt.axvline(x=5.7, color='black', linestyle='--', alpha=0.9)  # Entre Mayo y Junio
    plt.axvline(x=8.7, color='black', linestyle='--', alpha=0.9)  # Entre Agosto y Septiembre
    plt.axvline(x=11.7, color='black', linestyle='--', alpha=0.9)  # Entre Noviembre y Diciembre
    
    # Etiquetas de estaciones
    plt.text(2, maximo + 2 , 'Verano', color='black', ha='right')
    plt.text(4.3, maximo + 2 , 'Otoño', color='black', ha='center')
    plt.text(7.3, maximo + 2 , 'Invierno', color='black', ha='center')
    plt.text(10.3, maximo + 2 , 'Primavera', color='black', ha='center')
    
    # Establecer el límite y de los ejes y para controlar el valor máximo
    plt.ylim(0, maximo+10)
    
    # Mostrar el gráfico con leyenda fuera del plot
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Ajustar el diseño para evitar que la leyenda se recorte
    plt.tight_layout()
    plt.show()
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Supongamos que df1 y df2 son tus DataFrames de datos


# Cargar datos desde el archivo Excel
archivo_excel_coyhaique = "./CoyhaiqueV2.xlsx"  # Reemplaza con la ruta correcta de tu archivo Excel para Coyhaique
archivo_excel_valdivia = "./ValdiviaV2.xlsx"    # Reemplaza con la ruta correcta de tu archivo Excel para Valdivia

# Leer datos de los archivos Excel
df1 = pd.read_excel(archivo_excel_coyhaique)
df2 = pd.read_excel(archivo_excel_valdivia)

# Obtener la lista de delitos únicos
delitos = df1["delito"].unique()

# Crear un diccionario para almacenar las imágenes en miniatura y las funciones correspondientes
miniaturas = {}
funciones = {}

# Función para generar figuras de comparación para cada delito y devolver la miniatura
def generar_miniatura(delito, df1, df2, maximo):
    # Filtrar por delito especifico
    df_del1 = df1[df1["delito"]== delito]
    df_del2 = df2[df2["delito"]== delito]

    nombre1 = df_del1["nombre"].unique()
    nombre2 = df_del2["nombre"].unique()

    meses = df_del1["mes"].unique()
    
    # Ajustar las posiciones de las barras
    ancho_barra = 0.35

    pos_del1 = np.arange(len(meses))
    pos_del2 = pos_del1 + ancho_barra
    
    # Configurar el gráfico de barras agrupadas por mes
    fig, ax = plt.subplots()
    ax.bar(pos_del1, df_del1["tasa"], width=ancho_barra, label=f"{nombre1[0]}", color='gray', edgecolor='black')
    ax.bar(pos_del2, df_del2["tasa"], width=ancho_barra, label=f"{nombre2[0]}", alpha=0.7, color='lightgray', edgecolor='black')

    # Añadir etiquetas y título
    ax.set_xlabel("Mes")
    ax.set_ylabel("Tasa de Delito")
    ax.set_title(f"Comparación Mensual de la Tasa de {delito} entre {nombre1[0]} y {nombre2[0]}")
    ax.set_ylim(0, maximo+10)
    
    # Añadir líneas verticales para identificar las estaciones con etiquetas
    ax.axvline(x=2.7, color='black', linestyle='--', alpha=0.9)  # Entre Febrero y Marzo
    ax.axvline(x=5.7, color='black', linestyle='--', alpha=0.9)  # Entre Mayo y Junio
    ax.axvline(x=8.7, color='black', linestyle='--', alpha=0.9)  # Entre Agosto y Septiembre
    ax.axvline(x=11.7, color='black', linestyle='--', alpha=0.9)  # Entre Noviembre y Diciembre
    
    # Etiquetas de estaciones
    ax.text(2, maximo + 2 , 'Verano', color='black', ha='right')
    ax.text(4.3, maximo + 2 , 'Otoño', color='black', ha='center')
    ax.text(7.3, maximo + 2 , 'Invierno', color='black', ha='center')
    ax.text(10.3, maximo + 2 , 'Primavera', color='black', ha='center')
    
    # Mostrar el gráfico con leyenda fuera del plot
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Ajustar el diseño para evitar que la leyenda se recorte
    plt.tight_layout()
    
    # Guardar la figura en un archivo temporal
    ruta_temporal = f"./miniatura_{delito}.png"
    fig.savefig(ruta_temporal, bbox_inches='tight')
    plt.close(fig)  # Cerrar la figura para liberar recursos
    
    return ruta_temporal

# Crear miniaturas y funciones para cada delito
for delito in delitos:
    miniatura = generar_miniatura(delito, df1, df2, 50)  # 50 es un valor arbitrario para maximo
    miniaturas[delito] = miniatura
    
    # Función para mostrar la figura completa al seleccionar la miniatura
    def mostrar_funcion(delito_seleccionado):
        display(Image(filename=miniaturas[delito_seleccionado]))
    
    funciones[delito] = mostrar_funcion

# Crear widgets para seleccionar el delito y mostrar la miniatura correspondiente
dropdown_delitos = widgets.Dropdown(options=delitos, description='Selecciona un delito:')

# Asociar la función correspondiente al cambio en la selección del delito
widgets.interactive(funciones[delitos[0]], delito_seleccionado=dropdown_delitos)
```
